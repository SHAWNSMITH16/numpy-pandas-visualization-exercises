12. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.


```python
def get_db_url(db):
    from env import host, username, password
    '''
    return a formatted string containing username, password
    host and database for connecting to the mySQL server and the
    database indicated.
    
    env_file checks to see if the env.py exists in cwd
    
    '''
    username, password, host = (username, password, host)
    
    return f"mysql+pymysql://{username}:{password}@{host}/{db}"
```

```python
url_chipotle = get_db_url("chipotle")
query_chipotle = '''
SELECT
    *
FROM orders;
'''
df_chipotle = pd.read_sql(query_chipotle, url_chipotle)
df_chipotle
```

Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each

```python
df_chipotle['item_price'] = df_chipotle['item_price'].str.replace('$', '').astype(float)
```

```python
popular = df_chipotle.groupby('item_name')['item_price'].count().nlargest(4)
popular
```

```python
revenue = df_chipotle.groupby('item_name')['item_price'].sum().nlargest(10)
revenue
```

```python
sns.barplot(x = popular, y = revenue)
```
