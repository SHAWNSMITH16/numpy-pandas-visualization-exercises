1. Use matplotlib to plot the following equation:

y = x**2 − x + 2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = range(-100, 100)

#reads as: (equation) for every number in x
y = [(num ** 2 - num + 2) for num in x]

plt.plot(x, y)

plt.title('$y = x^2 - x + 2$')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0, 0), xytext = (0, 3000), arrowprops = {})
```

```python
x = np.linspace(-4, 4, 50)
y = x ** 2 - x + 2

plt.xlim(-5, 5)
plt.ylim(-1, 25)
plt.annotate('0, 0 point', xy=(0, 0), xytext=(-3, 1),
             arrowprops={'facecolor': 'darkgreen'})

plt.plot(x,y)
plt.show()
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

- y=√x
- y=x**3
- y=2x
- y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
n_rows = 2
n_cols = 2

# some data to play with
x = list(range(0, 100))


# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
y = [n for n in np.sqrt(x)]
plt.plot(x, y)
plt.title('sqrt of x')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
y = [x ** 3 for x in x]
plt.xlim(0, 20)
plt.ylim(0, 4000)
plt.plot(x, y)
plt.title('x cubed')


plt.subplot(n_rows, n_cols, 3)
y = [n**2 for n in x]
plt.plot(x, y)
plt.title('x squared')

plt.subplot(n_rows, n_cols, 4)
y = [1/(x+1) for x in x]
plt.xlim(0, 25)
plt.plot(x, y)
plt.title('x descending')


plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = list(range(0, 100))
d = [n for n in np.sqrt(x)]
z = [x ** 3 for x in x]
a = [n ** 2 for n in x]
b = [1/(x+1) for x in x]

plt.xlim(-1, 5)
plt.ylim(-1, 5)

plt.plot(x, d, c = 'blue')
plt.plot(x, z, c = 'red')
plt.plot(x, a, c = 'hotpink')
plt.plot(x, b, c = 'orange')
plt.title('Combination')


plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(0, 5))
d = [n for n in np.sqrt(x)]
z = [x ** 3 for x in x]
a = [n for n in np.double(x)]
b = [1/(x+1) for x in x]

plt.xlim(-1, 4)
plt.ylim(-1, 4)




plt.plot(x, d, c = 'blue', label = '$\sqrt{x}$')
plt.plot(x, z, c = 'red', label = '$x^3$')
plt.plot(x, a, c = 'hotpink', label = '${x}doubled$')
plt.plot(x, b, c = 'orange', label = '${x}descending$')
plt.title("Variances of 'x'")

plt.legend(loc='upper left')

plt.show()
```
